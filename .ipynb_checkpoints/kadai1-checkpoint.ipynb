# 単回帰分析の実装

## データの中心化

```python
# データの概要を表示
df.describe()
```

```python
df.mean()
```

```python
# 中心化
df_c = df - df.mean()
```

```python
df_c.head(3)
```

```python
df_c.describe()
```

```python
x = df_c['x']
y = df_c['y']
```

```python
plt.scatter(x,y)
plt.show()
```

## パラメータaの計算

```python
xx = x * x #*は要素積
```

```python
xy = x * y
```

```python
a = xy.sum() / xx.sum()
```

```python
a
```

## プロットして確認
$$
\hat{y} = ax
$$

```python
plt.scatter(x,y,label = 'y') # 実測値
plt.plot(x,a*x,label ='y_hat',color ='red') # 予測値
plt.legend() # 凡例の表示
plt.show()
```

## 予測値の計算
$$
\begin{align}
y - {\bar y} &= a(x-{\bar x})\\
\rightarrow y &= a(x-{\bar x}) + {\bar y}
\end{align}
$$

```python
x_new = 40 # 40平米の部屋
```

```python
mean = df.mean()
```

```python
mean['x']
```

```python
# 中心化
xc = x_new - mean['x']
```

```python
xc
```

```python
# 単回帰分析による予測
yc = a * xc
```

```python
xc
```

```python
#元のスケールの予測値
y_hat = a * xc + mean['y']
```

```python
y_hat
```

## 予測値を計算する関数の作成

```python
def predict(x):
    #定数項
    a = 10069.022519284063
    xm = 37.62222
    ym = 121065.0
    # 中心化
    xc = x - xm
    # 予測値の計算
    y_hat = a * xc + ym
    # 出力
    return y_hat
```

```python
#予測値
predict(40)
```

```python
predict(30)
```

```python
predict(50)
```
