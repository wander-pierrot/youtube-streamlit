## 
$$
\begin{align}
y - {\bar y} &= a(x-{\bar x})\\
\rightarrow y &= a(x-{\bar x}) + {\bar y}
\end{align}
$$

```python
x_new = 40 # 40
```

```python
mean = df.mean()
```

```python
mean['x']
```

```python
# 
xc = x_new - mean['x']
```

```python
xc
```

```python
# 
yc = a * xc
```

```python
def predict(x):
    #
    a = 10069.022519284063
    xm = 37.62222
    ym = 121065.0
    # 
    xc = x - xm
    # 
    y_hat = a * xc + ym
    # 
    return y_hat
```
